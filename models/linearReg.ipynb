```python
import numpy as np
import pandas as pd
import plotly.express as px
from datetime import datetime
import matplotlib.pyplot as plt
import mplfinance as mpf
import plotly.graph_objects as go
```

```python
nifty50_filepath =r"..\data\NIFTY_50_Historical_PR_01011990to14012024.csv" 
niftybank_filepath =r"..\data\NIFTY BANK_Historical_PR_01011990to14012024.csv" 
niftytotal_filepath =r"..\data\NIFTY TOTAL MARKET_Historical_PR_01011990to14012024.csv" 
niftyfmcg_filepath =r"..\data\NIFTY FMCG_Historical_PR_01011990to14012024.csv" 
```

```python
nifty50_df=pd.read_csv(nifty50_filepath,index_col=0,usecols=(1,2,3,4,5))
nifty50_df.index=pd.to_datetime(nifty50_df.index)
nifty50_df.fillna("",inplace=True)
nifty50_df
```

```python
niftybank_df=pd.read_csv(niftybank_filepath,index_col=0,usecols=(1,2,3,4,5))
niftybank_df.index=pd.to_datetime(niftybank_df.index)
niftybank_df.fillna("",inplace=True)
niftybank_df
```

```python
niftyfmcg_df=pd.read_csv(niftyfmcg_filepath,index_col=0,usecols=(1,2,3,4,5))
niftyfmcg_df.index=pd.to_datetime(niftyfmcg_df.index)
niftyfmcg_df.replace("-","",inplace=True)
niftyfmcg_df.fillna("",inplace=True)
niftyfmcg_df
```
